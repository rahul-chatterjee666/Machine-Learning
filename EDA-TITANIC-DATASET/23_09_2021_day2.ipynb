# Pivot Table

# A Pivot Table is used to summarise, sort, reorganise, group, count, total or average data stored in a table. It allows us to transform columns into rows and rows into columns. It allows grouping by any field (column), and using advanced calculations on them.

```python
import numpy as np
```

```python
df.pivot_table(values="number_of_fires",index="state",aggfunc=np.mean)  #aggfunc can be np.sum or np.median anything
```

```python
df.pivot_table(values="number_of_fires",index="state",aggfunc=np.mean,margins=True)  
#margin=true will provide all the means at the bottom
```

```python
fires_per_state=df.pivot_table(values="number_of_fires",index="state",aggfunc=np.mean,margins=True)
max_val=fires_per_state["number_of_fires"].max()+10
fires_per_state.plot(kind='barh',title="brazile fires by state ",xlim=(0,max_val),legend=False)


"""
- 'line' : line plot (default)
    - 'bar' : vertical bar plot
    - 'barh' : horizontal bar plot
    - 'hist' : histogram
    - 'box' : boxplot
    - 'kde' : Kernel Density Estimation plot
    - 'density' : same as 'kde'
    - 'area' : area plot
    - 'pie' : pie plot
    - 'scatter' : scatter plot (DataFrame only)
    - 'hexbin' : hexbin plot (DataFrame only)
"""
```

# using customized agg function 

```python
def max_minus_mean(group_data):
    return(group_data.max()-group_data.mean())
df.groupby('state')["number_of_fires"].agg(max_minus_mean)
```

# Adding A series in a dataframe

```python
print("Before putting into dataframe: ",type(df.groupby('state')["number_of_fires"].agg(max_minus_mean)))

print("after: ",type(pd.DataFrame(df.groupby('state')["number_of_fires"].agg(max_minus_mean).reset_index()))

pd.DataFrame(df.groupby('state')["number_of_fires"].agg(max_minus_mean).reset_index())
```

# Pandas merge

# Pandas has full-featured, high performance in-memory join operations idiomatically very similar to relational databases like SQL:

- **one-to-one joins:** for example when joining two DataFrame objects on their indexes (which must contain unique values).
- **many-to-one joins:** for example when joining an index (unique) to one or more columns in a different DataFrame.
- **many-to-many** joins: joining columns on columns.

![alt text](https://i.stack.imgur.com/hMKKt.jpg)

# 

```python
dfA = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

dfB = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                       'C': ['C0', 'C1', 'C2', 'C3'],
                       'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
dfA
```

```python
dfB
```

```python
# Merging the two dataset on keys  
pd.merge(dfA,dfB,on='key')
```

```python
# WE dont to append and give the NaN thing so we merge 

dfA.append(dfB)
```

```python
dfA = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

dfB = pd.DataFrame({'key': ['K4', 'K1', 'K2', 'K3'],
                       'C': ['C0', 'C1', 'C2', 'C3'],
                       'D': ['D0', 'D1', 'D2', 'D3']})
 
#  pd.merge(dfA, dfB, how='left', on='key')
```

```python
# here dfB is the big one and dfA is the smaller one means right join

pd.merge(dfA, dfB, how='right', on='key')  
```

```python
# here dfA is the big one and dfB is the smaller one means left join

pd.merge(dfA, dfB, how='left', on='key') # here dfA is the big one and dfB is the smaller one means left join
```

```python
# example of inner join

pd.merge(dfA, dfB, how='inner', on='key') # example of inner join
```

```python
 # full outer join
pd.merge(dfA, dfB, how='outer', on='key')
```

# ANOTHER JOIN BEHAVIOUR

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                      'B': ['B0', 'B1', 'B2']},index=['K0', 'K1', 'K2'])

left
```

```python
right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                       'D': ['D0', 'D2', 'D3']},index=['K0', 'K2', 'K3'])
                      

right
```

```python
left.join(right)    # it is classic example of left outer join
```
